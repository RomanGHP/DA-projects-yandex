<h1>Проект: Построение дашбордов в Tableau на основе данных о конференциях TED.</h1>

**Описание проекта:**

TED (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы. 

В разное время на TED-конференциях выступали математик Бенуа Мандельброт, теоретик искусственного интеллекта Марвин Минский, спортсменка Дана Ньяд и основатель Google Ларри Пейдж. В истории TED также были неоднозначные и даже скандальные выступления. Например, в 2010 году на конференции выступил Рэнди Пауэлл с рассказом о псевдонаучной «вихревой математике», а в 2014 году в конференции TEDMED участвовала Элизабет Холмс — основательница печально известного стартапа Theranos.

**Цель проекта:**

Провести комплексный анализ данных о выступлениях TED с помощью Tableau. Проект позволит получить ценную информацию о трендах развития TED, а также выделить наиболее популярные темы и форматы выступлений, которые привлекают внимание аудитории.

**Презентация доступна по ссылке:**

https://public.tableau.com/app/profile/roman.borshchinov/viz/TEDprojectpracticum/TEDproject

**Общие выводы:**

**1. История выступлений**

С 2008 года количество выступлений на TED-конференциях выросло в значительной степени. В исследовании мы изучили период с 2008 по 2021 год. 
* Большинство выступлений проходило в Соединенных Штатах Америки, за ними следуют Великобритания и Канада.
* Самые востребованные темы для выступлений касаются культуры, глобальных вопросов, науки, общества и технологий.
* В 2020 и 2021 годах наблюдался повышенный интерес к вопросам глобального характера, социальным проблемам и технологиям.

**2. Тематики выступлений**

* Наиболее популярные категории выступлений на TED – это наука, культура и технологии. 
* Распределение тематик выступлений в США, Великобритании и Канаде похожи. В Великобритании больше внимания уделяется глобальным вопросам, чем в США и Канаде.
* Тематики искусства, общества и образования, чаще всего получают аплодисменты. Меньше всего аплодисментов получатют спикеры по темам - дизайн, наука и глобальные вопросы. Больше всех аплодировали спикеру Сecile Richards c выступлением The political progress women have made and what's next.
* Чаще всего вызывают смех у зрителей темы культуры и дизайна. Меньше смешных моментов наблюдается в выступлениях, посвященных науке, обществу и глобальным вопросам. Самое смешное выступление у Maz Jobrani - A Saudi, an Indian and an Iranian walk into a Qatari bar.
* Зависимость между длительностью выступления и количеством просмотров не наблюдается. Самое просматриваемое выступление - Your body language may shape who you are, собрало 63,6 млн. просмотров. Самое длительное выступление - 3 secrets to Netflix success, длилось 63 минуты.

**3. Авторы выступлений**

* Среди авторов TED преобладают представители творческих профессий: писатели, журналисты и художники. 
* В основном, на одного автора приходится 1 выступление. Чаще всех выступал Hans Rosling (global health expert) - 8 раз.
* Дизайнер Tom Wujec выступал на TED 4 раза. Его выступление - Learn to use 13th-century astrolabe, собрало наименьшее количество просмотров среди всех его выступлений - 940542.

**4. Тематики категории «Другие»**

* Из категорий, не вошедших в топ-5 самых популярных тематик TED, наиболее часто встречается тема социальных изменений (social change).
* В 2020 году, из-за пандемии COVID-19, количество выступлений на TED значительно сократилось. Накануне пандемии, тема здравоохранения часто поднималась на  выступленях.
* С 2008 года, помимо топ-5, наиболее часто затрагиваются темы социальных изменений (social change), искусства (art), развлечений (entertainment), эффективности (perfomance) и здоровья (health). По этим темам регулярно проводились выступления практически каждый год.
